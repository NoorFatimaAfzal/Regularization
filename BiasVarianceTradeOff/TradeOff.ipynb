# Bias-Variance Trade-Off

## Introduction

The **bias-variance trade-off** is a fundamental concept in machine learning that describes the tension between the error introduced by the bias and the variance in a model. Understanding and balancing this trade-off is crucial for building models that generalize well to new data.

## Explanation

The goal is to find a model that appropriately balances bias and variance, minimizing the total error. A model with high bias pays little attention to the training data, resulting in underfitting. A model with high variance pays too much attention to the training data, resulting in overfitting. The bias-variance trade-off represents the balance between these two sources of error.

### Key Points:
- **Bias**: Error due to overly simplistic assumptions in the learning algorithm.
- **Variance**: Error due to excessive sensitivity to small fluctuations in the training set.
- **Trade-Off**: Balancing bias and variance to minimize the total error and achieve good generalization.

## Visualization

A common way to visualize the bias-variance trade-off is through the error decomposition:

<!-- ![Bias-Variance Trade-Off](https://upload.wikimedia.org/wikipedia/commons/6/67/Bias-variance_tradeoff.svg) -->

- **High Bias, Low Variance**: Simple models (e.g., linear regression with few features) that underfit the data.
- **Low Bias, High Variance**: Complex models (e.g., high-degree polynomial regression) that overfit the data.
- **Optimal Balance**: The model with the right complexity that minimizes total error.

## Examples

### Example 1: Linear vs. Polynomial Regression

- **Linear Regression**: May exhibit high bias and low variance.
- **Polynomial Regression**: As the degree increases, bias decreases and variance increases.



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

# Generate non-linear data
np.random.seed(0)
X = np.sort(5 * np.random.rand(80, 1), axis=0)
y = np.sin(X).ravel() + np.random.normal(0, 0.1, X.shape[0])

# Fit models
linear_model = LinearRegression()
linear_model.fit(X, y)
y_pred_linear = linear_model.predict(X)

poly_model = make_pipeline(PolynomialFeatures(degree=15), LinearRegression())
poly_model.fit(X, y)
y_pred_poly = poly_model.predict(X)

# Plot data and models
plt.scatter(X, y, color='black', label='Data')
plt.plot(X, y_pred_linear, color='blue', linewidth=2, label='Linear Model')
plt.plot(X, y_pred_poly, color='red', linewidth=2, label='Polynomial Model (degree=15)')
plt.title('linear model shows high bias and the polynomial model with degree 15 shows high variance')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
